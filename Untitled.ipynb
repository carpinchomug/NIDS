```python
import pandas as pd
import networkx as nx
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import TargetEncoder, StandardScaler, LabelEncoder, OrdinalEncoder
```

```python
file_name = "data/NF-UNSW-NB15-v2/data/NF-UNSW-NB15-v2.csv"
data = pd.read_csv(file_name)
```

```python
data.info()
```

```python
data = data.drop(columns=["L4_SRC_PORT", "L4_DST_PORT", "Label"])
data = data.groupby(by='Attack').sample(frac=0.01, random_state=42)
```

```python
# I'm going to drop categorical features for now. Once I package cateogry_encoders and I'll play with this data.
categorical_features = [
    "PROTOCOL",
    "TCP_FLAGS",
    "CLIENT_TCP_FLAGS",
    "SERVER_TCP_FLAGS",
    "ICMP_TYPE",
    "ICMP_IPV4_TYPE",
    "DNS_QUERY_ID",
    "DNS_QUERY_TYPE",
    "FTP_COMMAND_RET_CODE",
]

X = data.drop(columns=["Attack"] + categorical_features)
y = data[["Attack"]]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=13, stratify=y
)
```

```python
label_encoder = LabelEncoder().fit(y_train["Attack"])

y_train["Attack"] = label_encoder.transform(y_train["Attack"])
y_test["Attack"] = label_encoder.transform(y_test["Attack"])

numerical_features = list(set(X_train.columns[2:]).difference(set(categorical_features)))

sd = StandardScaler().fit(X_train[numerical_features])

X_train[numerical_features] = sd.transform(X_train[numerical_features])
X_test[numerical_features] = sd.transform(X_test[numerical_features])

# numeric_transformer = Pipeline(
#     steps = [("scaler", StandardScaler())]
# )

# categorical_transformer = Pipeline(
#     steps = [("encoder", OrdinalEncoder())]
# )

# preprocessor = ColumnTransformer(
#     transformers = [
#         ("numerical", numeric_transformer, numerical_features),
#         ("categorical", categorical_transformer, categorical_features),
#     ]
# )
```

```python
Xg_train = nx.from_pandas_edgelist(X_train, "IPV4_SRC_ADDR", "IPV4_DST_ADDR", numerical_features)
Xg_train = Xg_train.to_directed()
```
